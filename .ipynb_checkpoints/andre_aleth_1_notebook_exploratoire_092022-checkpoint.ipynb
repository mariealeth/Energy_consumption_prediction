# Functions

```python
def shapes(dataframes, dataframe_name):
    nb_rows = [data.shape[0] for data in dataframes]
    nb_columns= [data.shape[1] for data in dataframes]
    shapes = pd.DataFrame(columns=dataframe_name, data=[nb_rows, nb_columns], index=['nd_rows', 'nb_columns'])
    return shapes
```

```python
def number_of_null(dataframes, names):
    number_of_null = pd.DataFrame((dataframe.isna().sum() for dataframe in dataframes), index=names)
    return number_of_null
datas = [data16, data17, data19, data20]
```

```python
def suppress_rows(dataframe, column, values):
    '''Input : column = name of column, values = list'''
    '''Output : new dataframe'''
    mask = dataframe[column].isin(values)
    dataframe1 = dataframe[~mask]
    return dataframe1
```

```python
def suppress_rows_if_null(dataframe, column):
    '''Input : column = name of column'''
    '''Output : new dataframe'''
    mask = dataframe[column].isna()
    dataframe1 = dataframe[~mask]
    return dataframe1
```
